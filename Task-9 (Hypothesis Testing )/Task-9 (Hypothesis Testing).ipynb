## Task-9  Hypothesis Testing

<p>Q-1: Pista House selling Hyderabadi Chicken Dum biryani claims that each parcel packet has 500 grams of biryani (also mentioned on the label of packet). You are sceptic of their claims and believe that on average each packet does not contain 500 grams of biryani. How do you prove your claim? </p>


```python
#import important libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import statistics
from scipy.stats import t
```

```python
l = [490, 220, 470, 500, 495, 496, 496, 498, 508, 480]
sum(l)/len(l)
```

```python
# t_critical
alpha=1-0.95
t_critical=t.ppf(1-alpha/2,df=19)
print(t_critical)
```

```python
def t_score(n,sample_mean,pop_mean,s):
    num=sample_mean-pop_mean
    demo=s/n**0.5
    return num/demo
```

```python
s=statistics.stdev(l)
print(s)
```

```python
# t_score
s=statistics.stdev(l)
n=10
sample_mean=465.3
pop_mean=500
t_sco=t_score(n,sample_mean,pop_mean,s)
print(t_sco)
```

```python
# Ploting the sampling distribution with rejection regions

# Defining the x minimum and x maximum
x_min = 400
x_max = 600

# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = s / n**0.5

# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)

# Computing the left and right critical values (Two tailed Test)
t_critical_left = pop_mean + (-t_critical * std)
t_critical_right = pop_mean + (t_critical * std)

# Shading the left rejection region
x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='orange')

# Shading the right rejection region
x2 = np.linspace(t_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color='orange')

# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.0007))
```

```python
# Conclusion using t test

if(np.abs(t_sco) > t_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
p_value = 2 * (1.0 - norm.cdf(np.abs(t_sco)))

print("p_value = ", p_value)

if(p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

<p>Q-2: You have developed a new Natural Language Processing Algorithms and done a user study. You claim that the average rating given by the users is greater than 4 on a scale of 1 to 5. How do you prove this to your client? </p>

```python
# t_critical
alpha=1-0.95
t_critical=t.ppf(1-alpha,df=19)
print(t_critical)
```

```python
dta=[4,3,5,4,5,3,5,5,4,2,4,5,5,4,4,5,4,5,4,5]
statistics.stdev(dta)
```

```python
# Defining the sample and population parameters
n = 20
sample_mean = 4.25
pop_mean = 4
s = 0.85 
```

```python
t_sco2 = t_score(n, sample_mean, pop_mean,s)
print(t_sco2)
```

```python
# Ploting the sampling distribution with rejection regions

# Defining the x minimum and x maximum
x_min = 3.5
x_max = 4.5


# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = s / (n**0.5)


# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)


# Computing the right critical value (Right tailed Test)
t_critical_right = pop_mean + (t_critical * std)


# Shading the right rejection region
x1 = np.linspace(t_critical_right, x_max, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='orange')


# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.1))

# In this case sample mean falls in the rejection region
# i.e. Reject the Null Hypothesis
```

```python
# Conclusion using t test

if(t_sco2 > t_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
# Conclusion using p test

p_value = 1.0 - norm.cdf(np.abs(t_sco2))

print("p_value = ", p_value)

if(p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```
