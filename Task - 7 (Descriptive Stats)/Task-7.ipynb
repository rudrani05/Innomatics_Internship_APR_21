### 1.1 Mode

### 1.2 Median
<p>The sample median is the middle element of a sorted dataset. The dataset can be sorted in increasing or decreasing order. If the number of elements ùëõ of the dataset is odd, then the median is the value at the middle position: 0.5(ùëõ + 1). If ùëõ is even, then the median is the arithmetic mean of the two values in the middle, that is, the items at the positions 0.5ùëõ and 0.5ùëõ + 1. <br>

For example, if we have the data points 2, 4, 1, 8, and 9, then the median value is 4, which is in the middle of the sorted dataset (1, 2, 4, 8, 9). If the data points are 2, 4, 1, and 8, then the median is 3, which is the average of the two middle elements of the sorted sequence (2 and 4). The following figure illustrates this:</p>

<p> <img src='https://files.realpython.com/media/py-stats-04.f7b39a21dd2d.png' alt='Alt Text' title='Optional Title'></p>


#### Code Implementation
<p> We can calculate median using <i>median()</i> function:</p>

```python
print(df['Mthly_HH_Expense'].median()) # median of monthly expense
```

For example, in the set that contains the points 2, 3, 2, 8, and 12, the number 2 is the mode because it occurs twice, unlike the other items that occur only once.</p>

<p> The sample variance quantifies the spread of the data. It shows numerically how far the data points are from the mean. You can express the sample variance of the dataset ùë• with ùëõ elements mathematically as <br> ùë†¬≤ = Œ£·µ¢(ùë•·µ¢ ‚àí mean(ùë•))¬≤ / (ùëõ ‚àí 1), where ùëñ = 1, 2, ‚Ä¶, ùëõ <br>and mean(ùë•) is the sample mean of ùë•. </p>

### 3.Bivariate Analysis

#### Covariance
<p>In statistics, covariance is a measure of the association between variable X and Y. To be exact, it measures the linear relationship tendency of the variables.Here is the formula of covariance</p>
   
   $ cov_{x,y}=\frac{\sum(x_{i}-\bar{x})(y_{i}-\bar{y})}{N-1} $ <br>
$ cov_{x,y} $	=	covariance between variable a and y <br>
$x_{i}$	=	data value of x <br>
$y_{i}$	=	data value of y  <br>
$\bar{x}$	=	mean of x  <br>
$\bar{y}$	=	mean of y  <br>
N	=	number of data values


#### Correlation
<p> Pearson Correlation is one of the most used correlations during the data analysis process. Pearson correlation measures the linear relationship between variable continuous X and variable continuous Y and has a value between 1 and -1. In other words, the Pearson Correlation Coefficient measures the relationship between 2 variables via a line.</p>

$ \rho =\frac{\sum\left(x_{i}-\bar{x}\right)\left(y_{i}-\bar{y}\right)}{\sqrt{\sum\left(x_{i}-\bar{x}\right)^{2} \sum\left(y_{i}-\bar{y}\right)^{2}}} $<br>
r	=	correlation coefficient <br>
$x_{i}$	=	values of the x-variable in a sample <br>
$\bar{x}$	=	mean of the values of the x-variable <br>
$y_{i}$	=	values of the y-variable in a sample <br>
$\bar{y}$	=	mean of the values of the y-variable <br>
<p>When the correlation coefficient is closer to value 1, it means there is a positive relationship between variable X and Y. A positive relationship indicates an increase in one variable associated with an increase in the other. On the other hand, the closer correlation coefficient is to -1 would mean there is a negative relationship which is the increase in one variable would result in a decrease in the other. If X and Y are independent, then the correlation coefficient is close to 0 although the Pearson correlation can be small even if there is a strong relationship between two variables.</p>


#### Code Implementation

```python
print(df.corr(method='pearson'))
```

### 4. Normal Distribution
<p>The normal distribution is a form presenting data by arranging the probability distribution of each value in the data.Most values remain around the mean value making the arrangement symmetric.

We use various functions in numpy library to mathematically calculate the values for a normal distribution. Histograms are created over which we plot the probability distribution curve. </p>
#### Properties of Normal Distribution

We can simplify the Normal Distribution‚Äôs Probability Density by using only two parameters: ùùª Mean and ùõî2. This curve is symmetric around the Mean. Also as you can see for this distribution, the Mean, Median, and Mode are all the same.</p>

#### Empirical Rule for Normal Distribution

<p> <img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/Screenshot-from-2020-03-06-17-26-30.png' alt='Alt Text' title='Optional Title'></p>
<p>According to the Empirical Rule for Normal Distribution:<br>
<ol>
<li>68.27% of data lies within 1 standard deviation of the mean</li>
<li>95.45% of data lies within 2 standard deviations of the mean</li>
<li>99.73% of data lies within 3 standard deviations of the mean</li>
</ol>    
Thus, almost all the data lies within 3 standard deviations. This rule enables us to check for Outliers and is very helpful when determining the normality of any distribution.</p>

#### Skewed Distribution
<p>Normal Distribution is symmetric, which means its tails on one side are the mirror image of the other side. But this is not the case with most datasets. Generally, data points cluster on one side more than the other. We call these types of distributions <i>Skewed Distributions</i>. </p>

##### 1.Left Skewed Distribution
<p>When data points cluster on the right side of the distribution, then the tail would be longer on the left side. This is the property of Left Skewed Distribution. The tail is longer in the negative direction so we also call it <i>Negatively Skewed Distribution.</i> </p>

<p> <img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/Screenshot-from-2020-03-06-18-22-42.png' alt='Alt Text' title='Optional Title'></p>

<p>
In the Normal Distribution, Mean, Median and Mode are equal but in a negatively skewed distribution, we express the general relationship between the central tendency measured as:<br>
Mode > Median > Mean
 </p>
 
 ##### 2.Right Skewed Distribution
 <p>When data points cluster on the left side of the distribution, then the tail would be longer on the right side. This is the property of Right Skewed Distribution. Here, the tail is longer in the positive direction so we also call it <i>Positively Skewed Distribution.</i></p>
 
 <p> <img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/Screenshot-from-2020-03-12-14-02-09.png' alt='Alt Text' title='Optional Title'></p>
 
 <p>In a positively skewed distribution, we express the general relationship between the central tendency measures as:<br>
Mode < Median < Mean </p>

### 6. Box-Cox

<p>When you are dealing with real-world data, you are going to deal with features that are heavily skewed. Transformation technique is useful to stabilize variance, make the data more normal distribution-like, improve the validity of measures of association.
</p>
<p> <img src='https://miro.medium.com/max/856/1*Hk4xiGACg9Gtz2DwSERKqw.png' alt='Alt Text' title='Optional Title'></p>
<p> The problem with the Box-Cox Transformation is estimating lambda. This value will depend on the existing data, and should be considered when performing cross validation on out of sample datasets. Make sure to estimate lambda according to the training dataset.
</p>

#### Code Implementation

```python
# import modules
import numpy as np
from scipy import stats
from sklearn.model_selection import train_test_split
# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
# split into testing & training data
train,test = train_test_split(data, shuffle=False)

# transform training data & save lambda value
train_data,fitted_lambda = stats.boxcox(train)
# use lambda value to transform test data
test_data = stats.boxcox(test, fitted_lambda)
# (optional) plot train & test
fig, ax=plt.subplots(1,2)
sns.distplot(train_data, ax=ax[0])
sns.distplot(test_data, ax=ax[1])
```
